### Lab Report: Network and System Administration Practice (Exercise 02)

#### 1. Objective
The goal of this lab was to create a simple network service daemon and implement communication between a client and server. The tasks included:
- Creating a daemon in C that listens on a specific port.
- Connecting a client to the server and receiving a welcome message.
- Modifying the code to adapt to the required port and IP address.

---

#### 2. Environment
- **Server IP Address**: 192.168.5.100  
- **Client IP Address**: 192.168.5.102  
- **Port Number**: 14052  
- **Tools Used**: GCC compiler, terminal (Shell)

---

#### 3. Steps Performed

1. **Code Modification**
   - Changed the port number in both `server.c` and `client.c` to `14052`.
   - Set the destination IP address in `client.c` to `192.168.5.100`.

2. **Compilation**
   - Compiled the client code on the local machine:
     ```bash
     gcc -o client client.c
     ```
   - Compiled the server code on the server:
     ```bash
     gcc -o server server.c
     ```

3. **Server Execution**
   - Ran the server program on the server:
     ```bash
     ./server
     ```
   - The server started listening on port `14052`.

4. **Client Execution**
   - Ran the client program on the local machine:
     ```bash
     ./client
     ```
   - The client successfully connected to the server and received the welcome message:
     ```
     Received "Welcome to this wonderful server."
     ```

5. **Server Output**
   - The server output the following:
     ```
     Incoming connection from 192.168.5.102
     New message: "This is the client speaking !"
     Closing connection from 192.168.5.102
     ```

---

#### 4. Results
- A successful client-server communication was established.
- The server correctly listens on the specified port and responds to client requests.
- The client successfully receives the welcome message from the server.
- The server closes the connection properly after the client disconnects.

---

#### 5. Conclusion
This lab successfully implemented a basic TCP server-client communication model using C programming. Through this exercise, the fundamental steps of network programming, including port listening, data transmission, and connection closing, were understood. Further enhancements, such as multi-client support, configuration file reading, or UNIX socket integration, can be explored as per the lab instructions.
