```python
#Import goodbooks dataset with additional fields (see separate google colab notebook with the scraping code here : https://colab.research.google.com/drive/1AVD1QMp1hnqKmVjsVRUQ3HRUo4lqNBKP?usp=sharing
import pandas as pd
from ast import literal_eval

books_df = pd.read_csv('https://raw.githubusercontent.com/malcolmosh/MATH80629/main/books_enrichedv2.csv', index_col=[0], converters={"genres": literal_eval})

# head
display(books_df.head())
```

```python
# Check the data types of each column
dtypes = books_df.dtypes

# Print the data types
print(dtypes, "\n")

# Count the number of cells with inconsistent data
inconsistent_counts = {}
for col, dtype in dtypes.items():
    if dtype == 'object':  # Check for string columns
        inconsistent_counts[col] = books_df[col].str.contains('[^a-zA-Z0-9]').sum()
    elif dtype == 'int64':  # Check for integer columns
        inconsistent_counts[col] = books_df[col].apply(lambda x: not pd.api.types.is_integer(x)).sum()
    elif dtype == 'float64':  # Check for float columns
        inconsistent_counts[col] = books_df[col].apply(lambda x: not pd.api.types.is_float(x)).sum()

# Print the inconsistent counts
print(inconsistent_counts)
```

```python
import pandas as pd

# Read the CSV file
# books_df = pd.read_csv('archive/raw/1_book_details.csv')

# Check for null data
null_counts = books_df.isnull().sum()

# Print the null counts
print(null_counts)
```

```python
# Select the desired columns
desired_cols = ['title', 'description', 'genres', 'average_rating', 'authors']

# Create a new DataFrame that includes only the desired columns
new_df = books_df[desired_cols]

# Save the new DataFrame to a new CSV file
new_df.to_csv('archive/cleaned/1_git_books.csv', index=False)
```

```python
new_df = new_df.rename(columns={'title': 'title', 'description': 'description', 'genres': 'genre', 'average_rating': 'rating', 'authors': 'author'})
new_df = new_df[['title', 'author', 'description', 'genre', 'rating']]
# Save the new DataFrame to a new CSV file
new_df.to_csv('archive/cleaned/1_git_books.csv', index=False)
len(new_df.index)
```
