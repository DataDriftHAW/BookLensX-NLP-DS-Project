### Now we check if any of field has inconsistent or null data

```python
# Read the CSV file
df = pd.read_csv('archive/raw/2_book1-100k.csv')

# Check the data types of each column
dtypes = df.dtypes

# Print the data types
print(dtypes, "\n")

# Count the number of cells with inconsistent data
inconsistent_counts = {}
for col, dtype in dtypes.items():
    if dtype == 'object':  # Check for string columns
        inconsistent_counts[col] = df[col].str.contains('[^a-zA-Z0-9]').sum()
    elif dtype == 'int64':  # Check for integer columns
        inconsistent_counts[col] = df[col].apply(lambda x: not pd.api.types.is_integer(x)).sum()
    elif dtype == 'float64':  # Check for float columns
        inconsistent_counts[col] = df[col].apply(lambda x: not pd.api.types.is_float(x)).sum()

# Print the inconsistent counts
print(inconsistent_counts)
```

### Now we are checking how many data fields are empty or contains null type data

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('archive/raw/2_book1-100k.csv')

# Check for null data
null_counts = df.isnull().sum()

# Print the null counts
print(null_counts)
```

### Now as we can see, the null values are contained in the fields which we won't be needing.
so, we will now discard those fields and clean the file

```python
# Read the CSV file
df = pd.read_csv('archive/raw/2_book1-100k.csv')

# Get the columns with null values
null_cols = df.columns[df.isnull().sum() > 0]

# Create a new DataFrame that excludes the null columns
new_df = df.drop(null_cols, axis=1)

# Save the new DataFrame to a new CSV file
new_df.to_csv('archive/cleaned/2_book1-100k.csv', index=False)

print("Done removing the fields.")
```
